#### Count the number of tweets by a given Twitter account.
Ask the user for the handle (@handle) of a twitter account. You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account.

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
# your code here
def user_html(URL):
    html = None
    while True:
        try:
            handle = input("Insert user @handle to scrap: ").strip()
            html = requests.get(URL+handle)
            if html.status_code == 200:
                break
            else:
                print("No user found, try again!")
        except:
            print("No user found, try again!")
    return html


html = user_html(url).content

soup = BeautifulSoup(html,"lxml")
tweet_profile = soup.find_all("span", attrs={"class": "ProfileNav-value"})
tweets = tweet_profile[0].get("data-count")
tweets
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
# your code here
html = requests.get(url)
print(html.status_code)
html = html.content
````

````python
soup = BeautifulSoup(html,"lxml")
boxes = soup.find_all("a",{"class":"link-box"})
names = [box.strong.string.strip() for box in boxes]
articles = [box.bdi.string.replace("\xa0","").replace("+","").strip() for box in boxes]
zipped = [i for i in zip(names,articles)]
zipped
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
html = requests.get(url)
print(html.status_code)
html = html.content
````

````python
soup = BeautifulSoup(html,"lxml")
table = soup.find_all("table", attrs={"class": "wikitable sortable"})
rows = table[1].find_all("tr")
clean_rows = [row.text.replace("\n","   ").strip().split("   ") for row in rows][1:11]

lang = pd.DataFrame(clean_rows,columns=["Rank", "Language", "Native Speakers",
                                        "Percentage of World Population"])
lang.set_index(["Rank"],inplace=True)
lang.head(10)
````
